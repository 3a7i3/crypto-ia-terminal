## 1. Recherche instantanée dans les données
Implémentons une fonction de recherche permettant de filtrer instantanément un DataFrame selon un mot-clé ou une valeur.

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
# Exemple de données
df = pd.DataFrame({
    'Actif': ['BTC', 'ETH', 'SOL', 'BNB'],
    'Volume': [1200, 800, 450, 600],
    'Sharpe': [1.8, 1.2, 0.9, 1.5]
})
search_box = widgets.Text(placeholder='Recherche actif ou valeur...')
output = widgets.Output()
def on_search_change(change):
    with output:
        output.clear_output()
        val = change['new']
        if val:
            mask = df.apply(lambda row: row.astype(str).str.contains(val, case=False).any(), axis=1)
            display(df[mask])
        else:
            display(df)
search_box.observe(on_search_change, names='value')
display(search_box, output)
output.clear_output()
display(df)
```
