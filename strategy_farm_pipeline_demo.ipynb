# Introduction

Ce notebook montre comment exécuter un pipeline complet de génération, évolution, robustesse et ranking de stratégies quantitatives à l'aide des modules du dossier `QUANT_CORE/strategy_farm`.

> **But :** Fournir un exemple reproductible et interactif pour tester et comprendre chaque étape du pipeline quantitatif.


## 1. Importer les bibliothèques nécessaires
Nous allons utiliser pandas, numpy, matplotlib et les modules du dossier strategy_farm.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../')  # Pour permettre l'import local si besoin
from QUANT_CORE.strategy_farm.nursery import Nursery
from QUANT_CORE.strategy_farm.prototype_lab import PrototypeLab
from QUANT_CORE.strategy_farm.evolution_lab import EvolutionLab
from QUANT_CORE.strategy_farm.boost_lab import BoostLab
from QUANT_CORE.strategy_farm.robustness_lab import RobustnessLab
from QUANT_CORE.strategy_farm.ranking_engine import RankingEngine
from QUANT_CORE.strategy_farm.tracking import StrategyTracker
from QUANT_CORE.strategy_farm.strategy_registry import StrategyRegistry
```

## 2. Charger les modules du pipeline
Initialisons les modules du pipeline strategy_farm.

```python
nursery = Nursery()
prototype = PrototypeLab()
evolution = EvolutionLab()
boost = BoostLab()
robustness = RobustnessLab()
ranking = RankingEngine()
tracker = StrategyTracker()
registry = StrategyRegistry()
```

## 3. Définir et configurer le scénario du pipeline
Nous allons générer des stratégies simulées et préparer un DataFrame de marché fictif pour illustrer le pipeline.

```python
# Générer des stratégies simulées
strategies = []
for i in range(10):
    strat = {
        "id": f"strat_{i}",
        "sharpe": 0.5 + 0.05 * i,
        "drawdown": 0.1 * i,
        "winrate": 0.4 + 0.01 * i,
        "logic": "AND",
        "conditions": [
            {"indicator": "close", "operator": ">", "threshold": 100 + i}
        ],
        "position_size": 0.1 + 0.01 * i
    }
    strategies.append(strat)

# DataFrame de marché fictif
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=30)
close = np.cumsum(np.random.randn(30)) + 100
volume = np.random.randint(1000, 2000, size=30)
df_market = pd.DataFrame({'date': dates, 'close': close, 'volume': volume})
df_market.set_index('date', inplace=True)
df_market.head()
```

## 4. Exécuter le pipeline complet
Nous allons enchaîner toutes les étapes du pipeline : filtrage, optimisation, évolution, amplification, robustesse, tracking, enregistrement et ranking.

```python
# Filtrage rapide
filtered = nursery.quick_filter(strategies)
# Optimisation (pass-through)
optimized = prototype.optimize(filtered)
# Évolution
Evolved = evolution.evolve(optimized)
# Amplification (pass-through)
boosted = boost.amplify(Evolved)
# Robustesse
robust = robustness.stress_test(boosted)
# Tracking et enregistrement
for s in robust:
    tracker.track(s["id"], s)
    registry.register(s)
# Ranking
ranking.rank(robust)
# Récupérer les scores de fitness
fitness = [s["fitness"] for s in robust]
robust
```

## 5. Analyser et afficher les résultats du pipeline
Visualisons la distribution des scores de fitness et affichons les stratégies les mieux classées.

```python
# Visualisation de la distribution des scores de fitness
plt.figure(figsize=(6, 3))
plt.bar([s["id"] for s in robust], fitness)
plt.title("Distribution des scores de fitness des stratégies")
plt.xlabel("ID stratégie")
plt.ylabel("Fitness")
plt.tight_layout()
plt.show()

# Affichage des 3 meilleures stratégies
sorted_robust = sorted(robust, key=lambda x: x["fitness"], reverse=True)
print("Top 3 stratégies :")
for s in sorted_robust[:3]:
    print(s)
```
